```python
dStr = urllib.request.urlopen('https://hk.finance.yahoo.com/q/cp?s=%5EDJI').read()
getdStr=dStr.decode() 
#在python 3中urllib.read()返回bytes对象而非str，语句功能是将dStr转换成str
#convert dStr into str, urllib.read() returns bytes objects instead of str
m = re.findall('<tr><td class="yfnc_tabledata1"><b><a href=".*?">(.*?)</a></b></td><td class="yfnc_tabledata1">(.*?)</td>.*?<b>(.*?)</b>.*?</tr>', getdStr)
if m:
    print(m)
    print('\n')
    print (len(m))
else:  
    print ('not match')
```

## 编程练习

```python
poem = "How many roads must a man walk down\n\
Before they call him a man\n\
How many seas must a white dove sail\n\
Before she sleeps in the sand\n\
How many times must the cannon balls fly\n\
Before they're forever banned\n\
The answer my friend is blowing in the wind\n\
The answer is blowing in the wind"

f = open('Blowing in the wind.txt','w')
f.write(poem)
f.close()
```

```python
f = open('Blowing in the wind.txt','r')
cPoem = f.readlines()
len(cPoem)
```

```python
cPoem.insert(0,"Blowin' in the wind\n")
f.close()
f = open('Blowing in the wind.txt','w')
f.writelines(cPoem)
f.close()
```
